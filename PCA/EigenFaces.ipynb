```python
from sklearn import datasets
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
```

```python
oliv = datasets.fetch_olivetti_faces()
print(oliv)
```

```python
data = oliv.data
target = oliv.target
images = oliv.images
```

```python
#plotting the images
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(oliv.images[i],cmap=plt.cm.bone)
plt.show()
```

```python
pca = PCA(n_components=123)
pca.fit(data)
transformed = pca.transform(data)
```

```python
pca.components_.shape #eigen vectors
```

```python
#retreiving the images
inverse = pca.inverse_transform(transformed)
inverse.shape
approx_images = inverse.reshape((400,64,64))
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(approx_images[i],cmap=plt.cm.bone)
plt.show()
```

```python
#eigen faces
eigenVec = pca.components_
eigenFaces = eigenVec.reshape((123,64,64))
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8,i+1)
    ax.imshow(eigenFaces[i],cmap=plt.cm.bone)
plt.show()
```
