```python
documents = []
for category in movie_reviews.categories():
    for fileids in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileids),category))
documents[0:10]
```

```python
#currently we have negative and positive documents seperately. We need to shuffle it
import random
random.shuffle(documents)

documents[0:5]
```

```python
#creating a lemmatizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
#create a list of stop words and punctuations
from nltk.corpus import stopwords
import string
stop = set(stopwords.words('english'))
punc = list(string.punctuation)
stop.update(punc)
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_words(words):
    output = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag(w)
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output.append(clean_word.lower())
    return output
```

```python
documents = [(clean_words(document),category)for document,category in documents]
```

```python
documents[0:5]
```

```python
training_doc = documents[0:1500]
testing_doc = documents[1500:]
len(training_doc),len(testing_doc)
```

```python
all_words = []
for doc in training_doc:
    all_words += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000) #most common 3000 words
features = [i[0] for i in common]
len(features)
```

```python
def get_feature_dict(words):
    feature_dict = {}
    word_set = set(words)
    for w in features:
        feature_dict[w] = w in word_set
    return feature_dict
```

```python
get_feature_dict(training_doc[0][0]) #checking whether the function is working as intended
```

```python
training_data = [(get_feature_dict(doc),category) for doc,category in training_doc]
testing_data = [(get_feature_dict(doc),category) for doc,category in testing_doc]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
clf = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(clf,testing_data)
```

```python
clf.most_informative_features(15)
```

# Using sklearn classifiers

To use sklearn classifiers, we need to convert the format of the training and testing data because for nltk's naive bayes the data was of format -> [(dictionary of features,category)].
The format that we require for sklearn is - -> [2D matrix of the features], [1D matrix for the target]
So, to do so, we don't actually need to convert the data, we can simply use nltk's inbuilt sklearnClassifier which is basically a dummy classifier that uses a sklearn classifier.

```python
#using support vector classifier
from sklearn.svm import SVC
svc = SVC()
```

```python
from nltk.classify.scikitlearn import SklearnClassifier
clf = SklearnClassifier(svc)
```

```python
clf.train(training_data)
```

```python
nltk.classify.accuracy(clf,testing_data)
```

```python
#using random forests
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
clf1 = SklearnClassifier(rfc)
clf1.train(training_data)
```

```python
nltk.classify.accuracy(clf1,testing_data)
```

```python
#using logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
lgr = LogisticRegression()
```

```python
clf2 = SklearnClassifier(lgr)
clf2.train(training_data)
```

```python
nltk.classify.accuracy(clf2,testing_data)
```
